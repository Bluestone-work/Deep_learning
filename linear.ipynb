```python
import torch
import random
from d2l import torch as d2l
def create_Data(w , b , num_e):
    "y = wx + b + "
    x = torch.normal(0 , 1 ,(num_e,len(w)))
    y = torch.matmul(x , w) + b
    #element-wisePyTorchtorch.mul *,Matrix productPyTorchtorch.matmul
    y += torch.normal(0 , 0.01 , y.shape)
    return x,y.reshape((-1,1))

true_w = torch.Tensor([2,-3,4])
true_b = 4.2
features , label = create_Data(true_w,true_b,1000)
print('features:',features[0],'\nlabel:',label[0])
d2l.set_figsize()
d2l.plt.scatter(features[:,1].detach().numpy(),
                label.detach().numpy(), 1)
```
